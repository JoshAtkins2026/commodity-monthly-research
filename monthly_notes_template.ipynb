```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

# REPORT DATE
report_date = datetime.today().strftime('%B %Y')
print(f'Monthly Report: {report_date}')
```

```python
# LOAD DATA
df = pd.read_csv('commodity_data.csv', index_col='Date', parse_dates=True)
spread = df['Brent'] - df['WTI']

# KEY METRICS
current_spread = spread.iloc[-1]
spread_mean_30 = spread.rolling(30).mean().iloc[-1]
spread_std_30 = spread.rolling(30).std().iloc[-1]
zscore = (current_spread - spread_mean_30) / spread_std_30
# DRAWDOWN FROM 2024
spread_2024 = spread['2024-01-01':]
rolling_max_2024 = spread_2024.cummax()
current_drawdown_2024 = (spread_2024 - rolling_max_2024).iloc[-1]
# VOLATILITY
vol_30 = spread.rolling(30).std().iloc[-1]
vol_90 = spread.rolling(90).std().iloc[-1]

# DRAWDOWN
rolling_max = spread.cummax()
current_drawdown = (spread - rolling_max).iloc[-1]

# DAILY CHANGES
daily_change = df.iloc[-1] - df.iloc[-2]
daily_change_pct = ((df.iloc[-1] - df.iloc[-2]) / df.iloc[-2] * 100).round(2)

print('Data loaded successfully')
```

```python
# PRINT MONTHLY REPORT
print('=' * 60)
print(f'COMMODITY MARKETS MONTHLY REPORT — {report_date}')
print('=' * 60)

print('\n📊 BRENT-WTI SPREAD')
print(f'  Current Spread:        ${current_spread:.2f}')
print(f'  30 Day Average:        ${spread_mean_30:.2f}')
print(f'  Current Drawdown:      ${current_drawdown_2024:.2f}')
print(f'  Z Score:               {zscore:.2f}')
print(f'  Signal:                {"Buy" if zscore < -2 else "Sell" if zscore > 2 else "Neutral"}')

print('\n📈 VOLATILITY')
print(f'  30 Day Volatility:     ${vol_30:.2f}')
print(f'  90 Day Volatility:     ${vol_90:.2f}')
print(f'  Regime:                {"High" if vol_30 > vol_90 * 1.2 else "Low" if vol_30 < vol_90 * 0.8 else "Normal"}')

print('\n💹 DAILY PERFORMANCE')
for commodity in df.columns:
    change = daily_change_pct[commodity]
    arrow = '▲' if change > 0 else '▼'
    print(f'  {commodity:<12} {arrow} {change:>6.2f}%')

notes = f"""
{'=' * 60}
ANALYST NOTES — {report_date}
{'=' * 60}

MARKET OVERVIEW:
  - Major macro events this month?
  - Energy market response?
  - Geopolitical developments?

SPREAD ANALYSIS:
  - Spread vs 30 day average: ${current_spread:.2f} vs ${spread_mean_30:.2f}
  - Z Score signal: {"Buy" if zscore < -2 else "Sell" if zscore > 2 else "Neutral"}
  - What is driving the spread?

VOLATILITY:
  - Current regime: {"High" if vol_30 > vol_90 * 1.2 else "Low" if vol_30 < vol_90 * 0.8 else "Normal"}
  - Any spikes this month?

FORWARD CURVE:
  - Contango or backwardation?
  - Curve shape vs last month?

KEY RISKS:
  - What could move the spread significantly?
  - Scheduled events to watch?

OUTLOOK:
  - Base case next 30 days?
  - Key levels to watch?
{'=' * 60}
"""

print(notes)
```
